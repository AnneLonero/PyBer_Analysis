# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to check these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Challenge Deliverable 1. Generate a Ride-Sharing DataFrame by City Type

```python
#  1. Get the total rides for each city type
ride_count_by_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
ride_count_by_city_type.head()
```

```python
# 2. Get the total drivers for each city type
total_driver_by_city_type = city_data_df.groupby(["type"]).sum()["driver_count"]
total_driver_by_city_type.head()
```

```python
#  3. Get the total amount of fares for each city type
total_fares_by_city_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_by_city_type.head()
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_per_ride_by_city_type = total_fares_by_city_type/ride_count_by_city_type
average_fare_per_ride_by_city_type.head()
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver_by_city_type = total_fares_by_city_type/total_driver_by_city_type
average_fare_per_driver_by_city_type.head()
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({'Total Rides': total_driver_by_city_type,
                                 'Total Drivers':total_driver_by_city_type,
                                 'Total Fares':total_fares_by_city_type,
                                 'Average Fare per Ride':average_fare_per_ride_by_city_type,
                                 'Average Fare per Driver':average_fare_per_driver_by_city_type})
pyber_summary_df.head()
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
format_pyber_summary = {'Total Rides':"{:,.0f}", 'Total Drivers': "{:,.0f}", 'Total Fares':'${0:,.2f}', 'Average Fare per Ride':'${0:,.2f}', 'Average Fare per Driver':'${0:,.2f}'}
pyber_summary_df = pyber_summary_df.style.format(format_pyber_summary)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# Print the merged DataFrame for reference.
pyber_data_df.head()
```

```python
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
sum_fare_by_city_type_and_date_df = pd.DataFrame(pyber_data_df.groupby(['type', 'date']).sum()["fare"])
sum_fare_by_city_type_and_date_df.head()
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
sum_fare_by_city_type_and_date_df = sum_fare_by_city_type_and_date_df.reset_index()
sum_fare_by_city_type_and_date_df
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
sum_fare_pivot = sum_fare_by_city_type_and_date_df.pivot_table(index='date', columns='type', values='fare')
sum_fare_pivot
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-28'.
fare_date_type_jan_apr =sum_fare_pivot.loc[(sum_fare_pivot.index >= "2019-01-01") & 
                                (sum_fare_pivot.index <= "2019-04-29")]
fare_date_type_jan_apr                                 
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
fare_date_type_jan_apr.index = pd.to_datetime(fare_date_type_jan_apr.index)
fare_date_type_jan_apr
```

```python
# 6. Check that the datatype for the index is datetime using df.info()
fare_date_type_jan_apr.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
fare_week_type_jan_apr = fare_date_type_jan_apr.resample("W").sum()
fare_week_type_jan_apr
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(15,6))
fare_week_type_jan_apr.plot(kind="line", ax=ax, color=["lightskyblue", "coral", "gold"])
ax.set_title("Total Fare by City Type")
ax.set_ylabel("Fare ($USD)")
ax.set_xlabel(" ")
lgnd = plt.legend(fontsize="12", mode="Expanded", loc=(0.8,0.5), title="City Types")

# Save Figure
plt.savefig("analysis/PyBer_fare_summary.png")

plt.show()
```
